```python
url='https://timesofindia.indiatimes.com/entertainment/top-rated-movies/hindi/best-movies/2020/2742919'
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
html = requests.get(url).text
```

```python
soup = BeautifulSoup(html,'html.parser')
```

```python
crit_rat = []
for i in soup.find_all('div',class_='FIL_right'):
    crit_rat.append(i.find_all(class_='cricrating')[0].text.strip())
```

```python
Title= []
for i in soup.find_all('div',class_='FIL_right'):
    Title.append(i.find_all(['h3'])[0].text.strip())
```

```python
pd.DataFrame({'Title':Title,'Rating':crit_rat})
```
